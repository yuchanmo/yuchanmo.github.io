```
import dlib
sample = imgs[0]
```

```
detector = dlib.get_frontal_face_detector()
```

```
def detectitonFaces(sample:np.ndarray,display:bool=False):
  img_tmp = sample.copy()
  faces = detector(img_tmp)
  for f in faces:
    cv2.rectangle(img_tmp,(f.left(),f.top()),(f.right(),f.bottom()),(0,0,255),2)
  if display:
    cv2_imshow(img_tmp)
  return faces
                
```

```
for i in imgs[:10]:
  detectitonFaces(i,True)
```
